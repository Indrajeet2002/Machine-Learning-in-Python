Import the following libraries. numpy handles arrays, matplotlib is for visualization, pandas is for working with the matrix of features

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

First read the data. Then divide it into X and y vectors, which are the independent and dependent variables.

```python
dataset = pd.read_csv('/content/drive/My Drive/Data Preprocessing/Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

impute is a module in sklearn that is used for missing values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

The new means replaced the "nan" values from before

```python
print(X)
```

This code block encodes the independent variables from index 0 using OneHotEncoder

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

This code block uses the sklearn.preprocessing library's LabelEncoder to encode the dependent variables

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

This code block splits the data into training and testing sets, and uses four features X_train, X_test, y_train and y_test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

This code block takes the mean of the data and sets it to 0, and takes the standard deviation and sets it to 1

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```

```python
print(X_train)
```

```python
print(X_test)
```
