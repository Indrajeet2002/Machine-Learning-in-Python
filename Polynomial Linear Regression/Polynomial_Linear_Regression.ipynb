Import the numpy, matplotlib and pandas libraries to work with arrays, visualization and matrices, respectively

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Import the dataset and split into two vectors for independent and dependent variables

```python
dataset = pd.read_csv('/content/drive/My Drive/Polynomial Linear Regression/Position_Salaries_poly.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
print(X)
print(y)
```

Using LinearRegression to train the model on the entire dataset instead of dividing it into train and test

```python
from  sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

To train the entire dataset using the polynomial regression model, perform the transformation from X to X^4 as part of the preprocessing step. Use the degree parameter in PolynomialFeatures() to specify the power of X, and call the fit_transform() function. Finally, save that into a new X_poly variable and use it to fit the model.

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

Make a scatterplot to visualize the Linear Regression Results, and use X. Be sure to add the title and labels.

```python
plt.scatter(X, y, color = 'blue')
prediction_lin = lin_reg.predict(X)
plt.plot(X, prediction_lin, color = 'green')
plt.title('Truth of Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

Similarly, make a scatterplot to visualize the Polynomial Regression Results. This time use poly_reg since it has been trained with X^4.

```python
plt.scatter(X, y, color = 'blue')
X_poly = poly_reg.fit_transform(X)
prediction_poly = lin_reg_2.predict(X_poly)
plt.plot(X, prediction_poly, color = 'green')
plt.title('Truth of Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

Input a value of 5.5 into the linear regression model to determine how close the output is to the expected value

```python
lin_reg.predict([[5.5]])
```

Input the same value into the polynomial linear regression model to compare the output to that of the linear regression model

```python
lin_reg_2.predict(poly_reg.fit_transform([[5.5]]))
```
